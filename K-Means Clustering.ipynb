```python
df = df.drop(['level_0', 'index'], axis = 1)
df
```

#### Function to assign a Starting Pass -  X & Y location 

```python
def get_x(location):
    return location[0]

df['x'] = df['location'].apply(lambda x:f"{get_x(x)}").astype(float)


def get_y(location):
    return location[1]

df['y'] = df['location'].apply(lambda x:f"{get_y(x)}").astype(float)
```

#### Function to assign a Ending Passing - X & Y location

```python
def get_x(location):
    return location[0]

df['end_x'] = df['pass_end_location'].apply(lambda x:f"{get_x(x)}").astype(float)


def get_y(location):
    return location[1]

df['end_y'] = df['pass_end_location'].apply(lambda x:f"{get_y(x)}").astype(float)
```

```python
### Drop location and pass _end_location
df = df.drop(['location', 'pass_end_location'], axis = 1)
df
```

## IV - Build the Model

Assign k to the important features

```python
k = df.iloc[:,3:].values
k
```

### Elbow Method

```python
#We are plotting Within Cluster Sum Of Squares (WCSS) against the the number of clusters (K Value) to figure out the
#optimal number of clusters value.

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(k)
    wcss.append(kmeans.inertia_)
```

```python
#plot the line
plt.plot(range(1, 11), wcss, linewidth = 2, color = 'red', marker = '8')

#plot title and labels
plt.title('The Elbow Method', fontsize = 15, color = 'Blue')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 


plt.show()
```

The optimal K-value is found to be 5

### Train the Model with a cluster of 5

```python
kmeans_new = KMeans(n_clusters = 5, random_state = 0).fit(df.iloc[:,3:])
kmeans_new
```

### Centroids

```python
kmeans_new.cluster_centers_
```

### Predicted Cluster

```python
y_kmeans = kmeans_new.predict(df.iloc[:,3:])
y_kmeans[:25]
```
